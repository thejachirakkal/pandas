```python
schema=pd.read_csv("survey_results_schema.csv")
schema
```

```python
survey.columns
```

```python
survey.loc[(survey['ConvertedComp']>300000) &( survey['ConvertedComp']<1500000),['Country','Gender']]
```

# 1
1-age 20-30
2- salary-15L
3-Gender -Female

```python
val1=survey.loc[(survey['Age'].between(20,30)) & (survey['ConvertedComp']==1500000) & (survey['Gender']=='Woman')
           ,['Age','ConvertedComp','Gender']]
val1
```

```python
survey.value_counts('Gender')
```

```python
val1=survey.loc[(survey['Age'].between(20,30)) & (survey['ConvertedComp']==1500000) & (survey['Gender']=='Man')
           ,['Age','ConvertedComp','Gender']]
val1
```
